# Decoherence Simulation Notebook
This notebook simulates decoherence parameters for the Quantum-Secure Dark Net.

```python
import numpy as np
import matplotlib.pyplot as plt

g = 1e-12
m = 1e-22
Q = 1e9
eta = 0.999999

dist = np.logspace(0, 4, 200)
deco = g**2 * dist / (Q * eta)

plt.plot(dist, deco)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Distance (km)')
plt.ylabel('Decoherence Probability')
plt.title('Toy Decoherence Model')
plt.show()
```
